```python
# Print class properties
schema = client.schema.get()
print("Classes in the schema with properties:")
for class_info in schema['classes']:
    class_name = class_info['class']
    properties = class_info.get('properties', [])

    print(f"\nClass: {class_name}")
    if properties:
        print("  Properties:")
        for prop_details in properties:
            prop_name = prop_details['name']
            data_type = prop_details['dataType']
            print(f"    - {prop_name}: {data_type}")
    else:
        print("  No properties defined for this class.")
```
